```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler
import math
```

```python
data = df_freq.filter(['Close'])
X = data.values
train_data_len = math.ceil(len(X) * .8)
train_data_len
```

```python
# Data preparation using Min max scaler (Normalisation).
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(X)
```

```python
# Dividing the data into training-features and training-target values
window = 10
train_data = scaled_data[0: train_data_len, :]
x_train = []
y_train = []

for i in range(window, len(train_data)):
    x_train.append(train_data[i-window:i ,0])
    y_train.append(train_data[i, 0])
    if i <= window:
        print(x_train)
        print(y_train)
        print()
```

```python
import numpy as np
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
# Reshaping the data to feed in Neural Network.
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
x_train.shape
```

```python
# Importing keras for creating the model. 
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
# Model
model = Sequential()
model.add(LSTM(10, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(10, return_sequences=False, input_shape=(x_train.shape[1], 1)))
model.add(Dense(40))
model.add(Dense(1))
```

```python
# Compiling the model using Adam optimizer and calculating loss using MSE
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Fitting the data
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
# Testing the data
test_data = scaled_data[train_data_len-window : , :]
x_test = []
y_test = X[train_data_len:, :]
for i in range(window, len(test_data)):
    x_test.append(test_data[i-window:i ,0])
```

```python
x_test = np.asarray(x_test)
```

```python
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```python
# Predicitng future 
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
```

```python
predictions
```

```python
rmse = np.sqrt(np.mean(predictions - y_test) **2)
```
